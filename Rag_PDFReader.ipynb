```python
import chromadb

# Create a Chroma client
client = chromadb.Client()

# Specify the collection name or create a new one
collection_name = "my_collection"
collection = client.get_or_create_collection(name=collection_name)

# Assuming you have a list of embeddings and corresponding document IDs
embeddings = [
    # Example embeddings (Replace these with your actual embeddings)
    [0.1, 0.2, 0.3, 0.4],
    [0.2, 0.3, 0.4, 0.5],
    # Add more embeddings as needed
]

document_ids = [
    "doc1",
    "doc2",
    # Add more document IDs corresponding to your embeddings
]

# Optionally, you can add metadata (e.g., text content)
metadata = [
    {"text": "This is the first document."},
    {"text": "This is the second document."},
    # Add more metadata corresponding to your embeddings
]

# Upload embeddings to ChromaDB
collection.add(
    ids=document_ids,
    embeddings=embeddings,
    metadatas=metadata
)

print(f"Uploaded {len(embeddings)} embeddings to the '{collection_name}' collection.")
```
